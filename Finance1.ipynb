```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.ar_model import AutoReg
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/selvinfurtado01/fin_data_science/refs/heads/main/retvsind.csv")
```

```python
df.head()
```

```python
y = df.RET
x = df.sprtrn
x = sm.add_constant(x)
est = sm.OLS(y,x)
est = est.fit()
est.summary()
```

```python
dfstock = pd.read_csv("https://raw.githubusercontent.com/selvinfurtado01/fin_data_science/refs/heads/main/daily_stock_returns.csv")
dfindex = pd.read_csv("https://raw.githubusercontent.com/selvinfurtado01/fin_data_science/refs/heads/main/daily_index_returns.csv")
```

```python
dfindex.head()
```

```python
yindex = dfindex['vwretd']
modelindex = AutoReg(yindex, lags=5)
est = modelindex.fit()
print(est.summary())
```
