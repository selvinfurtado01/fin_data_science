```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
event_data = 'https://raw.githubusercontent.com/selvinfurtado01/fin_data_science/refs/heads/main/fin_data_science(event_study_dataset).csv'
t = pd.read_csv(event_data)
```

```python
t.head(10)
```

```python
scaar = []

for i in range(1,12):
    estimation_ret = t.iloc[0:45,i:i+1]
    estimation_mean = np.mean(estimation_ret)
    event_ret = t.iloc[45:50,i:i+1]
    ab = event_ret - estimation_mean
    caar = np.mean(ab)
    event_std = np.std(ab, ddof=1)
    scaar_event_std = caar / (event_std / np.sqrt(4))
    scaar.append(scaar_event_std)
```

```python
print(scaar)
```

```python
mean_scaar = np.mean(scaar)
std_scaar = np.std(scaar, ddof=1)
t = mean_scaar / (std_scaar / np.sqrt(11))
dof = 10

# p-value for 2-sided test
p_value = 2*stats.t.sf(abs(t), dof) 
print(p_value)
```
